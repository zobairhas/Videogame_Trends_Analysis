```python
# Regression is used to predict continuous variables. Assumes a few things:
# Null values are handled.
# Only numerical data is used.
# Values are scaled. In other words, data has been manipulated to ensure that the variance between the numbers won't skew results.
```

```python
# Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Read game_data csv as a DF
games_df = pd.read_csv(Path('Resources/game_data.csv'))
games_df
```

```python
games_df.dtypes
```

```python
# Chose metacritic_score as the independent variable
# because we assume recommendations to others
# depends on the metacritic_score 
plt.scatter(games_df.metacritic_score, games_df.recommendations)
plt.xlabel('metacritic_score')
plt.ylabel('recommendations')
plt.show()
```

```python
# Assigning X and y variables
# Unspecified rows and specified columns using reshape for X
X = games_df.metacritic_score.values.reshape(-1, 1)
y = games_df.recommendations
```

```python
X[:5]
```

```python
X.shape
```

```python
# Train/fit the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Create predictions
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
# Replot with red line
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
# Slope = coef
# Y intercept = intercept
print(model.coef_)
print(model.intercept_)
```
